```julia
include("DCSideBatteryModeling/src.jl")
```

```julia
parameters = get_params(Ub,fb,ωb,Sb,Vb)
```

```julia
ic = get_IC()
```

# Solving the system, pausing to introduce load step, and resuming simulation

```julia
tspan = (0.0,0.5)
prob = ODEProblem(ode_system!,ic.zero,tspan, parameters)
sol1 = solve(prob)
```

```julia
plot(sol1,vars=(0,13),title = "DC Voltage Before Load Step")
```

```julia
parameters.pl = 0.6;
```

```julia
tspan = (0.0,1)
prob = ODEProblem(ode_system!,sol1.u[end],tspan, parameters)
sol2 = solve(prob)
```

```julia
plot(sol2,vars=(0,13),title = "DC Voltage After Load Step")
```

# Using callbacks

```julia
function condition(u,t,integrator)
    t == 0.2
end
 
function affect!(integrator)
  parameters.pl = 0.6;
end
cb = DiscreteCallback(condition, affect!)
```

```julia
parameters = get_params(Ub,fb,ωb,Sb,Vb)
const tstop = [0.2]
```

```julia
parameters.pl
```

```julia
prob = ODEProblem(ode_system!,ic.zero,tspan, parameters)
sol3 = solve(prob,Tsit5(),callback = cb, tstops=tstop)
```
